## 아이리스 다중 분류

```python
import pandas as pd
df = pd.read_csv('data/iris.csv')
del df['Id']
df.head()
```

```python
dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]
Y_obj[0], Y_obj[50], Y_obj[100]
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
Y = e.fit_transform(Y_obj)
```

```python
Y
```

```python
import tensorflow as tf
from tensorflow import keras
# One hot encoding
Y_encoded = keras.utils.to_categorical(Y)
Y_encoded[0], Y_encoded[50], Y_encoded[100]
```

```python
Y_encoded.shape
```

```python
# seed 값 설정
seed = 0
tf.random.set_seed(seed)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = \
    train_test_split(X, Y_encoded, test_size=0.3, 
                     shuffle=True, random_state=seed)
```

```python
# 딥러닝 모델 설정
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(16, input_shape=(4,), activation='relu'),
    Dense(3, activation='softmax')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint
checkpointer = ModelCheckpoint(filepath="model/iris.hdf5", monitor='val_loss', 
                               verbose=1, save_best_only=True)
```

```python
# 모델 실행
model.fit(X_train, Y_train, epochs=100, batch_size=30, verbose=2,
          validation_data=(X_test, Y_test), callbacks=[checkpointer]) 
```

```python
from keras.models import load_model
del model
model = load_model("model/iris.hdf5")
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model.evaluate(X_test, Y_test, verbose=2)[1])) 
```

### Model 2

```python
# 딥러닝 모델 설정
model2 = Sequential([
    Dense(12, input_shape=(4,), activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')
])
model2.summary() 
```

```python
# 모델 컴파일 
model2.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

```python
checkpointer = ModelCheckpoint(filepath="model/iris2.hdf5", monitor='val_loss', 
                               verbose=1, save_best_only=True)
```

```python
# 모델 실행
model2.fit(X_train, Y_train, epochs=100, batch_size=30, verbose=2,
           validation_data=(X_test, Y_test), callbacks=[checkpointer]) 
```

```python
del model2
model2 = load_model("model/iris2.hdf5")
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model2.evaluate(X_test, Y_test, verbose=2)[1]))
```
